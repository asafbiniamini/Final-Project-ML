## Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split, KFold
from sklearn.datasets import fetch_openml
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from typing import List, Dict
%matplotlib inline
```

```python
# load
data = pd.read_csv("train.csv")
#Converting data to numpy arrays
X, y = data.values[:,:-1], data.values[:, -1]
print(f'X.shape: {X.shape}, y.shape: {y.shape}')
```

# Part 1: Explorization

```python
# Plot a bar chart of class distribution
class_counts = pd.Series(y).value_counts()
plt.bar(class_counts.index, class_counts.values)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Distribution')
plt.show()
```

```python
# Plot a histogram of a numeric feature
plt.hist(X[:, 2], bins=20)  # Replace 0 with the index of the feature you want to plot
plt.xlabel('Feature')
plt.ylabel('Count')
plt.title('Feature Distribution')
plt.show()
```

```python
# Separate the examples based on their labels
positive_examples = data[data['label'] == 1]
negative_examples = data[data['label'] == 0]

# Create the scatter plot for positive examples (label 1)
plt.figure(figsize=(5,3))
plt.scatter(positive_examples['vsize'], positive_examples['size'], color='red', label='Label 1')

# Set the labels and title for the positive examples plot
plt.xlabel('vsize')
plt.ylabel('size')
plt.title('Size vs. vsize Scatter Plot - malicious files')

# Show the plot for positive examples
plt.show()

# Create the scatter plot for negative examples (label 0)
plt.figure(figsize=(5,3))
plt.scatter(negative_examples['vsize'], negative_examples['size'], color='blue', label='Label 0')


# Set the labels and title for the negative examples plot
plt.xlabel('vsize')
plt.ylabel('size')
plt.title('Size vs. vsize Scatter Plot - benig files')

# Show the plot for negative examples

plt.show()
```

## Conclusion:

1. From the malicious: files with big size an small vsize more likely to be malicious

```python
#×¢×•×“ ×œ× × ×›×•×Ÿ


feature_values = X[:,4]
#data.iloc[:, 3].values

# Get the labels for the examples
#labels = data.iloc[:, -1].values

# Create separate datasets for labels 1 and 0
data_label_1 = feature_values[y == 1]
data_label_0 = feature_values[y == 0]

# Create the figure and axis for label 1
fig, ax_label_1 = plt.subplots()

# Set the axis limits and labels for label 1
ax_label_1.set_xlim(0, len(data_label_1) - 1)
ax_label_1.set_xlabel('Examples')
ax_label_1.set_ylabel('Feature')

# Plot the examples with label 1 as red dots
ax_label_1.plot(range(len(data_label_1)), data_label_1, 'ro')

# Create the figure and axis for label 0
fig, ax_label_0 = plt.subplots()

# Set the axis limits and labels for label 0
ax_label_0.set_xlim(0, len(data_label_0) - 1)
ax_label_0.set_xlabel('Examples')
ax_label_0.set_ylabel('Feature')

# Plot the examples with label 0 as blue dots
ax_label_0.plot(range(len(data_label_0)), data_label_0, 'bo')

# Show the graphs
plt.show()
```

```python
features = ['has_debug', 'has_relocations', 'has_resources', 'has_signature', 'has_tls']
data[features] = data[features].replace({None: 0})
data['Sum'] = data[features].sum(axis=1)

# Count the occurrences of each new feature value with label 1
result1 = data[data['label'] == 1]['Sum'].value_counts().sort_index()
result0 = data[data['label'] == 0]['Sum'].value_counts().sort_index()

# Create a new DataFrame with the results
table1 = pd.DataFrame({'Sum': result1.index, 'amount of examples with label 1': result1.values})
table0 = pd.DataFrame({'Sum': result0.index, 'amount of examples with label 0': result0.values})

# Print the table
print(table1)
print(table0)
```
