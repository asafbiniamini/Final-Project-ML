In this notebook we shall demonstrate simple usage of linear regression. First for regression, and then for classification using the prediction sign.

## Part 1: Linear regression on quantitative label

## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

### Load the diabetes dataset
Explanation of the data set can be found at https://www4.stat.ncsu.edu/~boos/var.select/diabetes.html
the features contain age, sex, bodi mass index (BMI), average blood pressure (BP), and 6 blood serum measurements.
The response of interest, a quantitative measure of disease progression one year after baseline.
Total of 442 patients.
Each column is mean centered and scaled to have variation of 1.

```python
diabetes_X_df, diabetes_y = datasets.load_diabetes(return_X_y=True, as_frame=True)
pd.concat((diabetes_X_df, diabetes_y), axis=1).head()
```

View stats on the target (a quantitative measure of disease progression one year after baseline)

```python
pd.Series(diabetes_y).describe()
```

### Use only one feature (the BMI)

```python
diabetes_X_one_feature = diabetes_X_df[['bmi']] # use [BMI] instead of BMI to keep it a DataFrame instead of a Series
print(f"diabetes_X_one_feature shape: {diabetes_X_one_feature.shape}")
diabetes_X_one_feature.head()
```

### Train

Notice the <b> intercept_ </b> attribute which gives us the bias ($w_0$) and <b> coef_ </b> attributes which gives us the coefficients (weights $w_1,...,w_{10}$)

```python
# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
# For linear regression, computes the w vector with the pseudo inverse times y, and saves it.
regr.fit(diabetes_X_one_feature, diabetes_y)

# Make predictions using the testing set
# For linear regression, simply computes Xw
diabetes_y_pred = regr.predict(diabetes_X_one_feature)

print('Intercept (w_0): ', regr.intercept_)
print('Coefficients (only w_1 in this case): ', regr.coef_)
print('Mean squared error: %.2f' % mean_squared_error(diabetes_y, diabetes_y_pred))
print(f'the trained model is (with rounded numbers):')
print(f'f(x)={regr.intercept_:.2f} + {regr.coef_[0]:.2f}*{diabetes_X_one_feature.columns[0]}')

# Plot outputs
plt.scatter(diabetes_X_one_feature, diabetes_y,  color='black')
plt.plot(diabetes_X_one_feature, diabetes_y_pred, color='blue', linewidth=3)
plt.xlabel(diabetes_X_one_feature.columns[0])
plt.ylabel(diabetes_y.name)
plt.title('Train predictions')

plt.show()
```

### Now lets use all features!
Do we expect to get a lower or a higher MSE?

```python
# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
# For linear regression, computes the w vector with the pseudo inverse times y, and saves it.
regr.fit(diabetes_X_df, diabetes_y)

# Make predictions using the testing set
# For linear regression, simply computes Xw
diabetes_y_pred = regr.predict(diabetes_X_df)

print('Intercept (w_0): ', regr.intercept_)
print('Coefficients(w_1,..,w_10): ', regr.coef_)
print('Mean squared error: %.2f' % mean_squared_error(diabetes_y, diabetes_y_pred))
# print('Coefficient of determination: %.2f' % r2_score(diabetes_y_train, diabetes_y_pred))

print(f'\nthe trained model is (with rounded numbers):')
print(f'f(x)={regr.intercept_:.2f} + ' + ' + '.join([f'{w:.2f}*{feature}' for w, feature in zip(regr.coef_, diabetes_X_df)]))

plt.show()
```

# Part 2: None linear transformation & linear classification

## Imports

```python
# the warning wrapper is to avoid an annoying warning of h5py
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import h5py  # use pip install h5py if an error raises
warnings.resetwarnings()
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from functools import reduce
from sklearn.linear_model import LinearRegression
```

### Load the usps dataset
Grayscale 16x16 digits (white is 255, black is 0).
The data is already scaled to 0-1 range (instead of 0-255).

```python
def hdf5(path, data_key = "data", target_key = "target", flatten = True):
    """
        loads data from hdf5:
        - hdf5 should have 'train' and 'test' groups
        - each group should have 'data' and 'target' dataset or spcify the key
        - flatten means to flatten images N * (C * H * W) as N * D array
    """
    with h5py.File(path, 'r') as hf:
        train = hf.get('train')
        X_train = train.get(data_key)[:]
        y_train = train.get(target_key)[:]
        if flatten:
            X_train = X_train.reshape(X_train.shape[0], reduce(lambda a, b: a * b, X_train.shape[1:]))
    return X_train, y_train

X, y = hdf5("usps.h5")
print("Dataset loaded. Shape: "+str(X.shape))
```

### View some samples

```python
num_samples_for_digit = 5 
num_classes = len(set(y))  # number of different digits

classes = set(y)
num_classes = len(classes)
fig, ax = plt.subplots(num_samples_for_digit, num_classes, sharex = True, sharey = True, \
                       figsize=(num_classes, num_samples_for_digit))

for label in range(num_classes):
    class_idxs = np.where(y == label)
    for i, idx in enumerate(np.random.randint(0, class_idxs[0].shape[0], num_samples_for_digit)):
        # the imshow with 'gray' option, recieves either 0-255 integers, or 0-1 floats.
        ax[i, label].imshow(X[class_idxs[0][idx]].reshape([16, 16]), 'gray')
        ax[i, label].set_axis_off()
plt.show()
```

# Preprocessing
The input dimension is too big!

Regression models will use 257 weights (256 pixel + bias).
Lets reduce the input dimension to 2: Symmetry, and intensity.
How many weights will our model have in this case?

```python
def symmetry(image):
    im = image.reshape((16,16))
    return np.mean(1-np.abs(im - im[::-1,:]))
def intensity(image):
    return np.mean(image)

def phi(image):
    #nonlinear transformation
    return np.array([intensity(image),symmetry(image)])

# apply phi on all the dataset
X_pro = np.array([phi(im) for im in X])

print('Done preprocessing. New shape: ', X_pro.shape)
```

## Binary classification between '1' and '5'
Now we shall only take the 1 and 5 digits, and try to classify them!

```python
X_1, X_pro_1 = X[y == 1], X_pro[y == 1]
X_5, X_pro_5 = X[y == 5], X_pro[y == 5]
```

## Plot

```python
plt.figure(figsize=(8,8))
plt.plot([im[0] for im in X_pro_5],[im[1] for im in X_pro_5],'ro',label='5', markeredgecolor='black', markeredgewidth=1)
plt.plot([im[0] for im in X_pro_1],[im[1] for im in X_pro_1],'bx',label='1')
plt.xlabel('Intensity')
plt.ylabel('Symmetry')
plt.legend(numpoints=1)
plt.show()
```

# Examine the data and plot outliers
It looks like there is an outlier '5' and several '1's. Lets look at the 5 with High Symmetry and low intensity, and on the 1 with the smallest symmetry.

```python
# we search in the processed data, but plot the unprocessed.
# the 'nearest' interpolation is used to turn off the interpolation.
figure = plt.figure(figsize=(12, 9))
ax = plt.subplot(1, 2, 1)
ax.set_title("5 digit with highest symmetry")
ax.imshow(X_5[np.argmax(X_pro_5[:,1])].reshape([16, 16]),'gray', interpolation='nearest')

ax = plt.subplot(1, 2, 2)
ax.set_title("1 digit with smallest symmetry")
ax.imshow(X_1[np.argmin(X_pro_1[:,1])].reshape([16, 16]),'gray', interpolation='nearest')  #1 with minimal Symmetry

plt.show()
```

## Classification time!
In order to classify using the prediction sign, we first need to change the labels to $\pm$1.
<br>
Therefore, we first change the label of 5 to -1

```python
# set the dataset with 1 label for 1 and -1 for 5
X_pro_1_5 = np.concatenate([X_pro_1,X_pro_5])
y_1_5 = np.ones(X_pro_1_5.shape[0])
y_1_5[-X_pro_5.shape[0]:] = -1

# train
regr = LinearRegression().fit(X_pro_1_5, y_1_5)
print('Intercept: ', regr.intercept_)
print('Coefficients: ', regr.coef_)
print('\nAfter defining labels {1: 1, 5: -1}:')
print(f'f(x)=sign({regr.intercept_:.2f} + {regr.coef_[0]:.2f}*Intensity + {regr.coef_[1]:.2f}*Symmetry)') 
# plot
# we take 2 points where w_0+w_1x_1+w_2x_2 = 0. Once we nullify x_1 and once x_2
#(0,a), (b,0)
a,b = -regr.intercept_/regr.coef_[1], -regr.intercept_/regr.coef_[0]
plt.figure(figsize=(8,8))
plt.plot([im[0] for im in X_pro_5],[im[1] for im in X_pro_5],'ro',label='5', markeredgecolor='black', markeredgewidth=1)
plt.plot([im[0] for im in X_pro_1],[im[1] for im in X_pro_1],'bx',label='1')
plt.plot([0,b],[a,0],'black',label='linear_regression')
plt.ylim(0.4,1) # since there arent any samples with lower symmetry
plt.xlim(0.03,0.56) # since there arent any samples with higher Intensity
plt.xlabel('Intensity')
plt.ylabel('Symmetry')
plt.legend(numpoints=1)#handles=[red_dot, blue_x])
plt.show()
```

The result is ok, but its not great. The line is drawn down a bit. This is due to the squared loss, which draws the classifier towards the samples which are far from it.
This is one of the side effects of using a regression model (with appropriate loss) for classification.
Its a nice initial benchmark, but we will meet better classifiers later on.

# Classification - try it yourself!

```python
digit = 1 # modify the digit as wanted
class_idxs = np.where(y == digit)
idx = class_idxs[0][np.random.randint(0, class_idxs[0].shape[0])]
image = X[idx]
print(f'digit index: {idx}')
print(f'digit label: {y[idx]}')
print('\nAfter defining labels {1: 1, 5: -1}:')
print(f'digit prediction: {np.sign(regr.predict([phi(image)])[0])}')                      
plt.imshow(image.reshape([16, 16]), 'gray')
plt.show()
```
